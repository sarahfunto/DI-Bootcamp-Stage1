```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#definir les paramètres
city = 10
months = 12
low_temperature = -5
high_temperature = 35
```

```python
#generer random
temperature_data = np.random.uniform(low=low_temperature, high=high_temperature, size=(city, months))
```

```python
# Definir nom des villes et chaque mois
city_names = [f'City_{i+1}' for i in range(city)]
month_names = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun',
               'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
```

```python
#creer un dataFrame
temperature_df = pd.DataFrame(data=temperature_data, index=city_names, columns=month_names)

temperature_df
```

```python
#Calculer la température annuelle moyenne par ville ---
annual_avg_temp = temperature_df.mean(axis=1)

annual_avg_temp
```

```python
# la ville la plus chaude
hottest_city = annual_avg_temp.idxmax()
hottest_city
```

```python
#la ville la plus froide
coldest_city = annual_avg_temp.idxmin()
coldest_city
```

```python
#Visualiser les villes par mois

plt.figure(figsize=(12, 6))
for city in temperature_df.index:
    plt.plot(month_names, temperature_df.loc[city], label=city)
plt.title('Temperatures pour 10 villes')
plt.xlabel('Mois')
plt.ylabel('Temperature (°C)')
plt.show()
```

```python
# temperature moyenne annuelle
annual_avg_temp = temperature_df.mean(axis=1)
sorted_avg_temp = annual_avg_temp.sort_values(ascending=False)
plt.figure(figsize=(10, 6))
plt.bar(sorted_avg_temp.index, sorted_avg_temp.values)
plt.title('Villes par température moyenne annuelle')
plt.xlabel('Villes')
plt.ylabel('Temperature moyenne annuelle (°C)')
plt.show()
```

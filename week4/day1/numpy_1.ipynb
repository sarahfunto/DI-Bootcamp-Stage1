### Why Numpy and not Regular Python lists? 

#### Modifying the numbers in a list 

```python
# Regular Python
python_list = [1, 2, 3, 4, 5]

for i in range(len(python_list)):
    python_list[i] = python_list[i] * 2

python_list
```

```python
python_list = [1, 2, 3, 4, 5] 

python_list_doubled = [x * 2 for x in python_list]

python_list_doubled
```

```python
sum(python_list_doubled)
```

```python
# Numpy Array
numpy_array = np.array([1, 2, 3, 4, 5])

doubled = numpy_array * 2

doubled
```

```python
doubled.sum() # numpy has its own mathematical methods
```

```python
import time 

size = 1_000_000
python_list = list(range(size))

start = time.time() # before running the doubling
result_list = [x * 2 for x in python_list]
python_time_list = time.time() - start # after running the doubling

python_time_list
```

```python
size = 1_000_000
numpy_array = np.arange(size)

start = time.time() # before running the doubling
result_numpy = numpy_array * 2
python_time_numpy = time.time() - start # after running the doubling

python_time_numpy
```

when numpy 
- working with numbers 
- need speed 
- doing linear algebra, statistics etc.
- pandas uses numpy by default for numerical columns

when lists
- working with mixed data types (strings , numbers , objects mixed together in a single place)
- need to append/remove items frequently

#### More Conditions for masks 

```python
array = np.arange(10)
array
```

```python
# get all even numbers
even_numbers_mask = array % 2 == 0
```

```python
array[even_numbers_mask]
```

```python
# get numbers between 3 and 7 (3,4,5,6,7)
between_3_and_7_mask = (array >= 3) & (array <= 7)
```

```python
between_3_and_7_mask
```

```python
array[between_3_and_7_mask]
```

### 'Real World' Example: Filtering Temperature Data

```python
np.random.seed(0) # Seed makes sure that the random numbers generated WILL BE THE SAME EACH RUN 
temperatures = np.random.uniform(15, 35, 20)
temperatures
```

```python
hot_days_mask = temperatures > 30
```

```python
temperatures[hot_days_mask]
```

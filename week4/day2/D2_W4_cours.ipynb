### PERCENTILES

```python
The value below which a given percentage of dat falls
25th percentil(Q1) : 25% of data is below this
50th percentil(Q2) : 50% as median
75th percentil (Q3) : 75% od data os below this
Why it matters : 
- Understand the data distribution : see where most data is
- Detect outliers : values beyond 95th percentile
-Performence metrics : Top 10% of students
```

### Variance and standard Deviation

Probleme example
Group A 48, 49, 50, 51, 52 --> scores are close together
Group B : 20, 40, 50, 60, 80 --> score ar apart

```python
group_a = np.array([48, 49, 50, 51, 52])
group_b = np.array([20, 40, 50, 60, 80])

group_a.mean(), group_b.mean()
```

```python
#Variance
#calculatinf variance :
##a.finf how far each values od from th mean = calculate the distance
[2, 1, 0, 1, 2]
## 2. squares the distances (make tham positive)
[4, 1, 0, 1, 4]
##3. Average the squared distances
mean([4, 1, 0, 1, 4]) -> 2
small variance (2) - data points are close together
```

```python
#Standard Deviation

Square root of the Variance
Variance: 2_500_000_000 dollars^2
Standard deviation : 50_000 <-much clear!

house prices typically vary by +- 50_000 from the mean
```

```python
np.array([70, 75, 80, 85, 90)
mean : 80

[10, 5, 0, 5, 10]
>
[100, 25, 0, 25, 100]
>
50 (variance, not readable)
>
7(standard deviation , scores vary by +- 7 points)
```

```python
### Correlation

Measures how two varaibles move together.

    Correlation Coefficient (r)
- + 1.0 :  Perfect positive correlation(both inscrease together)
-   0.0 :  No correlation (independant vairables)
- - 1.0 : Perfect negative correlation (ont increases, other decreases)

Real world examples:
-Bill size vs tip amount : positive (bigger bill -> bigger tip)
-Shoe size vs QI : Zero (unrelated)
-Temperature vs heating cost : Negative (warmer -> less heating)
```

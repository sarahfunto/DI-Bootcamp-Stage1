### Variance and Standard Deviation 

Problem example:
Group A: 48, 49, 50, 51, 52 -> Scores are close together
Group B: 20, 40, 50, 60, 80 -> Scored are apart

```python
group_a = np.array([48, 49, 50, 51, 52])
group_b =  np.array([20, 40, 50, 60, 80])

group_a.mean(), group_b.mean()
```

```python
# Variance
## calculating variance:
## 1. Find how far each values if from the mean = calculate the distance
[2, 1, 0, 1, 2]
## 2. squares the distances (make them positive)
[4, 1, 0, 1, 4]
## 3. Average the squared distances
mean([4, 1, 0, 1, 4]) -> 2

small varaince (2) - data points are close together 
```

```python
# Standard Deviation 

Square root of the Variance

Variance: 2_500_000_000 dollars^2
Standard deviation: 50_000 <- much clearer!

house prices typically vary by +-50_000 from the mean
```

```python
np.array([70, 75, 80, 85, 90])
mean: 80

[10, 5, 0, 5, 10]
>
[100, 25, 0, 25, 100] 
>
50 (variance, not readable)
>
7 (Standard deviation, Scores vary by +- 7 points)
```

### Correlation

Measures how two variables move together. 

Correlation Coefficient (r):
-  + 1.0:  Perfect positive correlation (both increase together) 
-    0.0:  No correlation (independent variables)
-  - 1.0:  Perfect negative correlation (one increases, other decreases)
 
real world exampels:
- Bill size vs tip amount: Positive (bigger bill -> bigger tip)
- Shoe size vs IQ: Zero (unrelated)
- Temperature vs heating cost: Negative (warmer -> less heating)
